```python
import numpy as np
x = np.arange(0,100)
y = x * 2
z = x ** 2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("Summary")
ax.set_xlabel("Investment")
ax.set_ylabel("Profit")
ax.plot(x,z)
```

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
ax2 = fig1.add_axes([0.2,0.5, 0.2,0.2])
```

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
ax2 = fig1.add_axes([0.2,0.5, 0.2,0.2])
ax1.set_xlabel("Investment")
ax2.set_ylim(250,600)
ax2.set_xlim(20,22)
ax2.plot(x,z)
ax1.plot(x,z)
```

```python
#PArrellel Analysis
fig, axes = plt.subplots(1,2, figsize=(10,5))
axes[0].plot(x,y, color='red', lw=5, ls='--')
axes[1].plot(x,z, color='green', lw=5)
```
