```python
def labelencoder(dframe, col):
    le = preprocessing.LabelEncoder()
    le.fit(dframe[col])
    trans_data = le.transform(dframe[col])
    dframe[col]=trans_data
    return dframe
```

```python
df
```

```python
df = labelencoder(df,'Pass')
```

```python
df = labelencoder(df,'State')
```

```python
#OneHotEncoding
df = pd.read_csv('student_cleaned.csv')
df['State'] = ['AK', 'DP', 'AK','RP','RP','AK', 'DP']
df
```

```python
#onehot encoding function
def onehotencode_data(dframe, col):
    onehot = pd.get_dummies(dframe[col],prefix=col)
    dframe.drop(col, axis =1, inplace=True)
    dframe = dframe.join(onehot)
    return dframe
```

```python
df = onehotencode_data(df,'State')
print(df)
```

```python
#OneHotEncoding using sklearn
df = pd.read_csv('student_cleaned.csv')
df['State'] = ['AK', 'DP', 'AK','RP','RP','AK', 'DP']
df
```
